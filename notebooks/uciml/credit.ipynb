```python
import sys
from ucimlrepo import fetch_ucirepo

sys.path.append("..")


# fetch dataset
credit_approval = fetch_ucirepo(id=27)


# variable information
print(credit_approval.variables)
```

```python
cat_cols = ["A1", "A4", "A5", "A6", "A7", "A9", "A10", "A12", "A13"]
num_cols = ["A2", "A3", "A8", "A11", "A14", "A15"]
```

```python
dataset = credit_approval.data["original"]
dataset.head()
```

```python
from tools.auto_clf_bench import dataset_benchmark


benchmark = dataset_benchmark(
    dataset, "A16", cat_cols=cat_cols, num_cols=num_cols, n_trials=1
)
```

```python
benchmark
```

```python
import pandas as pd


benchmark_df = (
    pd.DataFrame(benchmark)
    .pivot_table(index="model", values="value", columns="metric")
    .sort_values("f1_score", ascending=False)
    .round(2)
)
```
