```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-1, 0, 1000)

# Compute the real and imaginary parts of f(z(x))
u = -1 - 2 * x
v = -2 * x - 2 * x**2

# Plot the image
plt.figure()
plt.plot(u, v, label="Image of D under f(z) = z^2")
plt.title("Image of D under f(z) = z^2")
plt.xlabel("Re(f(z))")
plt.ylabel("Im(f(z))")
plt.grid()
plt.axis("equal")
plt.legend()
plt.show()
```
